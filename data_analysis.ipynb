GroupBy in Pandas

```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore (1).csv')

impute = SimpleImputer(missing_values = np.nan , strategy = 'mean')
impute.fit(df.iloc[ : , 2:3 ].values)
df.iloc[ : , 2:3 ] = impute.transform(df.iloc[ : , 2:3 ].values)

df = df.dropna()
```

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]

df_pr.sort_values( by = 'Rating', ascending = True ).head()
```

```python
df.head()
```

Q-1 name the apps that are in ART_AND_DESIGN with rating more then 4.5 in ascending order WRT rating ?

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]

df_pr.sort_values( by = 'Rating', ascending = True ).head()
```

which category is habing maximum average rating descending order?

```python
df.groupby('Category').mean()['Rating'].sort_values(ascending=False)
```

Q-3 how many paid apps are there in each category in descending order?

```python
df.groupby('Category')
```

```python
df_pr = df[df['Type'] == 'Paid']

df_pr.groupby('Category').count()['Type'].sort_values(ascending = False)
```
